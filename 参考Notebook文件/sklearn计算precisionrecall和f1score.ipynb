```python
import numpy as np
from sklearn.metrics import precision_recall_fscore_support
y_true = np.array([0, 1, 2, 2, 1, 0, 0, 1]) #label实际值
y_pred = np.array([2, 1, 1, 2, 0, 0, 1, 2]) #label预测值
precisions, recalls, f1_scores, _ = precision_recall_fscore_support(y_true, y_pred, average=None, labels = range(3)) 
#直接调用scikit-learn函数可以返回precision，recall和f1score
print(precisions,recalls,f1_scores)  #分别打印precision，recall和f1score
```
