```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib
plt.rcParams['font.family'] = ['Microsoft YaHei', 'sans-serif'] # 保证显示中文，且正确显示负数
```

## **折线图**

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.title('折线图I')
plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)

plt.plot(x, y1, label="sin(x)", color="blue", linestyle="-.")
plt.plot(x, y2, label="cos(x)", color="red", linestyle="--")
plt.plot(x, y3, label="sin(x) + cos(x)", color="green", linestyle=":")

plt.title('折线图II', fontsize=16)
plt.xlabel('X轴', fontsize=14)
plt.ylabel('Y轴', fontsize=14)

plt.legend()

plt.show()
```

## **散点图**

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y)
plt.title('散点图')
plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.show()
```

## **柱状图**

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [5, 7, 3, 8, 4]

plt.bar(x, y)
plt.title('柱状图I')
plt.xlabel('类别')
plt.ylabel('值')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [5, 7, 3, 8, 4]
values2 = [3, 6, 2, 9, 5]

x = np.arange(len(categories))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, values1, width, label='数据组1', color='skyblue')
plt.bar(x + width/2, values2, width, label='数据组2', color='orange')

plt.title('柱状图II', fontsize=16)
plt.xlabel('类别', fontsize=14)
plt.ylabel('值', fontsize=14)
plt.xticks(x, categories)

for i, v in enumerate(values1):
    plt.text(i - width/2, v + 0.1, str(v), ha='center', fontsize=12)
for i, v in enumerate(values2):
    plt.text(i + width/2, v + 0.1, str(v), ha='center', fontsize=12)

plt.legend()

plt.show()
```

## **直方图**

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('随机数据的分布情况', fontsize=16)
plt.xlabel('数据值', fontsize=14)
plt.ylabel('频数', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.5)

mean_value = np.mean(data)
std_value = np.std(data)
plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label=f'均值: {mean_value:.2f}')
plt.axvline(mean_value - std_value, color='green', linestyle='--', linewidth=2, label=f'标准差: {std_value:.2f}')
plt.axvline(mean_value + std_value, color='green', linestyle='--', linewidth=2)

plt.legend()

plt.show()
```

## **饼图**

## **箱线图**

```python
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=2, scale=1.5, size=100)
data3 = np.random.normal(loc=-2, scale=0.5, size=100)
data = [data1, data2, data3]

plt.figure(figsize=(10, 6))
plt.boxplot(data, patch_artist=True, showmeans=True)

plt.title('随机数据的箱线图', fontsize=16)
plt.xlabel('分组', fontsize=14)
plt.ylabel('数据值', fontsize=14)

plt.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.text(1, 3.5, '数据组1', fontsize=12, color='blue')
plt.text(2, 5, '数据组2', fontsize=12, color='orange')
plt.text(3, -1, '数据组3', fontsize=12, color='green')

plt.show()
```

## **面积图**

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)

plt.figure(figsize=(10, 6))
plt.fill_between(x, y1, color="skyblue", alpha=0.4, label="sin(x)")
plt.fill_between(x, y2, color="orange", alpha=0.4, label="cos(x)")
plt.fill_between(x, y3, color="green", alpha=0.4, label="sin(x) + cos(x)")

plt.title('面积图', fontsize=16)
plt.xlabel('X轴', fontsize=14)
plt.ylabel('Y轴', fontsize=14)

plt.legend()

plt.show()
```

## **足球场**
请使用Matplotlib中的patches绘制

```python
import matplotlib.patches as patches

# 创建一个画布和子图
fig, ax = plt.subplots()

# 绘制足球场边界
rect = patches.Rectangle((-10, -5), 20, 10, linewidth=2, edgecolor='black', facecolor='none')
ax.add_patch(rect)

# 绘制中圈
circle = patches.Circle((0, 0), 2, linewidth=2, edgecolor='black', facecolor='none')
ax.add_patch(circle)

# 绘制球门
goal1 = patches.Rectangle((-10, -1.5), 0.5, 3, linewidth=2, edgecolor='black', facecolor='none')
ax.add_patch(goal1)
goal2 = patches.Rectangle((9.5, -1.5), 0.5, 3, linewidth=2, edgecolor='black', facecolor='none')
ax.add_patch(goal2)

# 绘制球门区
goal_area1 = patches.Rectangle((-10, -2.5), 2, 5, linewidth=2, edgecolor='black', facecolor='none')
ax.add_patch(goal_area1)
goal_area2 = patches.Rectangle((8, -2.5), 2, 5, linewidth=2, edgecolor='black', facecolor='none')
ax.add_patch(goal_area2)

# 绘制中线
ax.plot([0, 0], [-5, 5], color='black', linewidth=2)

# 设置坐标轴
plt.xlim(-11, 11)
plt.ylim(-6, 6)


# 显示图形
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')
plt.show()
```
